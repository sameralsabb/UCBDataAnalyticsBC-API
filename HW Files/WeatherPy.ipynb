```python
import gmaps
import pandas as pd
import random
import decimal
from citipy import citipy
import numpy as np
import requests
from pprint import pprint 

#Google developer API key
from google_api import api_key as gkey
#OpenWeather API key
from samer_apikey import api_key  
```

```python
lat = np.random.uniform(-90,90,[500,1])
lat
lng = np.random.uniform(-90,90,[500,1])
lng
lats_lngs = zip(lat,lng)
```

```python
df = pd.DataFrame({'lat': lat.ravel(), 'lng': lng.ravel()})
table = df.round(2)
table["city name"] = " "
table['country code'] = " "
table.head()
```

```python
# create a params dict
params = {"key": gkey}

for index,row in table.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    lat = row['lat']
    lng = row['lng']
    
    params['address'] = f"{lat},{lng}"
    
    cities_name = requests.get(base_url, params=params)
    
    cities_name = cities_name.json()
    
```

```python
pprint(cities_name)
```

```python
city_name = cities_name["results"][0]['address_components'][0]['long_name']
city_name
```

```python
cities = []
for row in lats_lngs:
#     row[0], row[1] = coordinate_pair
    cities.append(citipy.nearest_city(row[0], row[1]))
```
